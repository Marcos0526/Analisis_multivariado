<h2>$\color{blue}{\text{1.1 Descriptive Statistics of the data}}$</h2>
A large data set is bulky, and its very mass poses a serious obstacle to any attempt to visually extract pertinent information. Much of the information contained in the data can be assessed by calculating certain summary numbers, known as descriptive statistics. One relies most heavily on descriptive statistics that measure location, variation, and linear association.
    
$\color{red}{\text{Remember that...:}}$
Assuming that we have observed \(n\) realizations of a \(p\)-dimensional random variable. We arrange the data in an data matrix $\mathbf{X}(n\times p)$:

$$
\mathbf{X} =
\begin{pmatrix}
x_{11} & \cdots & x_{1p} \\
\vdots & \ddots & \vdots \\
\vdots &        & \vdots \\
x_{n1} & \cdots & x_{np}
\end{pmatrix}.
$$

The rows $\mathbf{x}_i = (x_{i1}, \dots, x_{ip}) \in \mathbb{R}^p$ denote the \(i\)-th observation of a \(p\)-dimensional random variable $\mathbf{X} \in \mathbb{R}^p$.

The sample mean can be computed from the \(n\) measurements on each of the $(p)$ variables, so that, in general, there will be $p$ sample means:
$$
\bar{x}_k = \frac{1}{n} \sum_{i=1}^{n} x_{ik}, 
\qquad k = 1, 2, \dots, p.
$$

The mean can be rewritten in matrix form by the vector of means, $\bar{\mathbf{x}}$, of the $p$ variables:

$$
\bar{\mathbf{x}} =
\begin{pmatrix}
\bar{x}_1 \\
\vdots \\
\bar{x}_p
\end{pmatrix}
= n^{-1}\mathbf{X}^\top \mathbf{1}_n .
$$

Here, $\mathbf{1}_n$ is the $n$-dimensional vector of ones.

A measure of spread is provided by the sample variance, defined for $(n)$ measurements on the first variable as

$$
s_1^2 = \frac{1}{n} \sum_{i=1}^{n} (x_{i1} - \bar{x}_1)^2,
$$

where $\bar{x}_1$ is the sample mean of the $x_{i1}$'s. In general, for $p$ variables, we have
$$
s_k^2 = \frac{1}{n} \sum_{i=1}^{n} (x_{ik} - \bar{x}_k)^2,
\qquad k = 1, 2, \dots, p.
$$

One uses the notation $s_{kk}$ to denote the variance computed from measurements on the $k$-th variable, and one has the notational identity
$$
s_k^2 = s_{kk}
$$

The square root of the sample variance, $s_k$, is known as the **sample standard deviation**. This measure of variation uses the same units as the observations. 
A measure of linear association between the measurements of variables is provided by the **sample covariance** or the average product of the deviations from their respective means. If large values for one variable tend to occur with large values for the other variable, and small values also occur together, then $s_{ij}$ will be positive. If large values of one variable occur with small values of the other variable, then $s_{ij}$ will be negative. If there is no particular association between the two variables, then $s_{ij}$ will be approximately zero.
In general, the **sample covariance** between the $i$-th and $k$-th variables is
$$
s_{ik} = \frac{1}{n} \sum_{j=1}^{n} (x_{ji} - \bar{x}_i)(x_{jk} - \bar{x}_k),
\qquad i = 1, 2, \dots, p,\quad k = 1, 2, \dots, p.
$$
Note that the covariance reduces to the sample variance when $i = k$. Moreover, $s_{ik} = s_{ki}$ for all $i$ and $k$.

This measure of the  dispersion of the $n$ observations can be characterized by the empirical covariance matrix of the $p$ variables:
$$
\mathbf{S}
= n^{-1}\mathbf{X}^\top \mathbf{X} - \bar{\mathbf{x}}\bar{\mathbf{x}}^\top
= n^{-1}\left(\mathbf{X}^\top \mathbf{X}
- n^{-1}\mathbf{X}^\top \mathbf{1}_n \mathbf{1}_n^\top \mathbf{X}\right).
$$
Equivalently, the covariance matrix can be written as
$$
\mathbf{S}
= \frac{1}{n} \sum_{i=1}^{n}
(\mathbf{x}_i - \bar{\mathbf{x}})
(\mathbf{x}_i - \bar{\mathbf{x}})^\top .
$$
which can also be rewritten as
$$
\mathbf{S}
= n^{-1}\mathbf{X}^\top \mathbf{H}\mathbf{X},
$$
where the **centering matrix** $\mathbf{H}$ is defined as
$$
\mathbf{H}
= \mathbf{I}_n - n^{-1}\mathbf{1}_n \mathbf{1}_n^\top .
$$
Note that the centering matrix is **symmetric** and **idempotent**. Indeed,
$$
\mathbf{H}^2
= \left(\mathbf{I}_n - n^{-1}\mathbf{1}_n\mathbf{1}_n^\top\right)
  \left(\mathbf{I}_n - n^{-1}\mathbf{1}_n\mathbf{1}_n^\top\right)
$$

$$
= \mathbf{I}_n 
- n^{-1}\mathbf{1}_n\mathbf{1}_n^\top 
- n^{-1}\mathbf{1}_n\mathbf{1}_n^\top
+ \left(n^{-1}\mathbf{1}_n\mathbf{1}_n^\top\right)
  \left(n^{-1}\mathbf{1}_n\mathbf{1}_n^\top\right)
$$

$$
= \mathbf{I}_n - n^{-1}\mathbf{1}_n\mathbf{1}_n^\top
= \mathbf{H}.
$$

As a consequence, the covariance matrix $\mathbf{S}$ is **positive semidefinite**, i.e.,

$$
\mathbf{S} \succeq 0.
$$

Indeed, for all $\mathbf{a} \in \mathbb{R}^p$,

$$
\mathbf{a}^\top \mathbf{S} \mathbf{a}
= n^{-1}\mathbf{a}^\top \mathbf{X}^\top \mathbf{H}\mathbf{X}\mathbf{a}
$$

$$
= n^{-1}\left(\mathbf{a}^\top \mathbf{X}^\top \mathbf{H}^\top\right)
   (\mathbf{H}\mathbf{X}\mathbf{a})
   \quad \text{since } \mathbf{H}^\top \mathbf{H} = \mathbf{H},
$$

$$
= n^{-1}\mathbf{y}^\top \mathbf{y}
= n^{-1} \sum_{j=1}^{p} y_j^2
\ge 0,
$$

where

$$
\mathbf{y} = \mathbf{H}\mathbf{X}\mathbf{a}.
$$

It is well known from the one-dimensional case that

$$
\frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

as an estimator of the variance exhibits a bias of order $n^{-1}$ (Breiman, 1973). In the multidimensional case,
$$
\mathbf{S}_u = \frac{n}{n-1}\mathbf{S}
$$

is an **unbiased estimator** of the true covariance matrix.

The final descriptive statistic considered is the **sample correlation coefficient** (or Pearson’s product–moment correlation coefficient). This measure of linear association between two variables does not depend on the units of measurement. The sample correlation coefficient for the $i$-th and $k$-th variables is defined as
$$
r_{ik}
= \frac{\sum_{j=1}^{n} (x_{ji} - \bar{x}_i)(x_{jk} - \bar{x}_k)}
{\sqrt{\sum_{j=1}^{n} (x_{ji} - \bar{x}_i)^2}
 \sqrt{\sum_{j=1}^{n} (x_{jk} - \bar{x}_k)^2}},
\qquad i = 1,2,\dots,p,\quad k = 1,2,\dots,p.
$$

Note that $r_{ik} = r_{ki}$ for all $i$ and $k$.

The sample correlation coefficient is a **standardized version of the sample covariance**, where the product of the square roots of the sample variances provides the standardization.  Observe that $r_{ik}$ has the same value whether $n$ or $n-1$ is chosen as the divisor for $s_{ii}$, $s_{kk}$, and $s_{ik}$.

The sample correlation coefficient $r_{ik}$ can also be viewed as a **sample covariance**. Suppose the original values $x_{ji}$ and $x_{jk}$ are replaced by standardized values

$$
\frac{x_{ji} - \bar{x}_i}{s_{ii}^{1/2}}
\quad \text{and} \quad
\frac{x_{jk} - \bar{x}_k}{s_{kk}^{1/2}}.
$$

The standardized values are commensurable because both sets are centered at zero and expressed in standard deviation units. The sample correlation coefficient is then the sample covariance of these standardized observations.

Although the signs of the sample correlation and the sample covariance are the same, the correlation is ordinarily easier to interpret because its magnitude is bounded. To summarize, the sample correlation coefficient $r$ has the following properties:

1. The value of $r$ satisfies
   $$
   -1 \le r \le 1.
   $$

2. The value of $r$ measures the **strength of linear association**.  
   If $r = 0$, there is no linear association between the components.  
   If $r < 0$, one variable tends to be larger when the other is smaller than its mean.  
   If $r > 0$, large values tend to occur together, and small values tend to occur together.

3. The value of $r_{ik}$ remains unchanged if the measurements of the $i$-th variable are transformed as
   $$
   y_{ji} = a x_{ji} + b, \qquad j = 1,2,\dots,n,
   $$
   and the $k$-th variable as
   $$
   y_{jk} = c x_{jk} + d, \qquad j = 1,2,\dots,n,
   $$
   provided that the constants \(a\) and \(c\) have the same sign.

   
Finally note that if

$$
\mathbf{D} = \mathrm{diag}(s_{ii}),
$$

then the **correlation matrix** is

$$
\mathbf{R}
= \mathbf{D}^{-1/2}\mathbf{S}\mathbf{D}^{-1/2},
$$

where $\mathbf{D}^{-1/2}$ is a diagonal matrix whose main diagonal entries are $(s_{ii})^{-1/2}$.

<h3>$\color{blue}{\text{Linear Transformation}}$</h3>

In many practical applications, we need to study **linear transformations** of the original data. This motivates the question of how to calculate summary statistics after such transformations. Let $\mathbf{A}$ be a $q \times p$ matrix and consider the transformed data matrix

$$
\mathbf{Y} = \mathbf{X}\mathbf{A}^\top 
= (\mathbf{y}_1, \dots, \mathbf{y}_n)^\top .
$$

The row  
$$
\mathbf{y}_i = (y_{i1}, \dots, y_{iq}) \in \mathbb{R}^q
$$
can be viewed as the \(i\)-th observation of a \(q\)-dimensional random variable  
$$
\mathbf{Y} = \mathbf{A}\mathbf{X}.
$$
In fact,
$$
\mathbf{y}_i = \mathbf{x}_i \mathbf{A}^\top .
$$

The mean vector and empirical covariance matrix of the transformed data are

$$
\bar{\mathbf{y}}
= \frac{1}{n}\mathbf{Y}^\top \mathbf{1}_n
= \frac{1}{n}\mathbf{A}\mathbf{X}^\top \mathbf{1}_n
= \mathbf{A}\bar{\mathbf{x}},
$$

$$
\mathbf{S}_Y
= \frac{1}{n}\mathbf{Y}^\top \mathbf{H}\mathbf{Y}
= \frac{1}{n}\mathbf{A}\mathbf{X}^\top \mathbf{H}\mathbf{X}\mathbf{A}^\top
= \mathbf{A}\mathbf{S}_X \mathbf{A}^\top .
$$

---

<h3>$\color{blue}{\text{Nonhomogeneous Linear Transformation}}$</h3>


If the transformation is **nonhomogeneous**, i.e.,

$$
\mathbf{y}_i = \mathbf{A}\mathbf{x}_i + \mathbf{b}, 
\qquad \mathbf{b} \in \mathbb{R}^{q},
$$

only the mean changes:

$$
\bar{\mathbf{y}} = \mathbf{A}\bar{\mathbf{x}} + \mathbf{b}.
$$

Special Case: $q = 1$

If $q = 1$, then

$$
y = \mathbf{X}\mathbf{a}, 
\qquad
y_i = \mathbf{a}^\top \mathbf{x}_i,
$$

and

$$
\bar{y} = \mathbf{a}^\top \bar{\mathbf{x}}, 
\qquad
S_y = \mathbf{a}^\top \mathbf{S}_X \mathbf{a}.
$$


<h3>$\color{blue}{\text{Mahalanobis Transformation}}$</h3>

A special case of a linear transformation is the **Mahalanobis transformation**:

$$
\mathbf{z}_i = \mathbf{S}^{-1/2} (\mathbf{x}_i - \bar{\mathbf{x}}),
\qquad i = 1, \dots, n.
$$

Let

$$
\mathbf{Z} = (\mathbf{z}_1, \dots, \mathbf{z}_n)^\top .
$$

Then the covariance matrix of the transformed data is

$$
\mathbf{S}_Z
= n^{-1}\mathbf{Z}^\top \mathbf{H}\mathbf{Z}
= \mathbf{I}_p.
$$

Thus, the **Mahalanobis transformation removes correlation** between variables and **standardizes each variance to one**.

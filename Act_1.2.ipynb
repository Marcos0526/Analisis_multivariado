```python
import os
import kagglehub

dir= "csvs"
os.environ["KAGGLEHUB_CACHE"] = os.path.abspath(dir)
path = kagglehub.dataset_download("marceaxl82/shopping-mall-paid-search-campaign-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('csvs/datasets/marceaxl82/shopping-mall-paid-search-campaign-dataset/versions/1/final_shop_6modata.csv')
df.head()
```

```python
df.info()
```

```python
df_num=df.select_dtypes(include=[np.number])
df_num.columns
```

---
### 1)


```python
mean_per_column = np.mean(df_num, axis=0)
print(f"Promedios por Columna", "\n", mean_per_column) 
```

---
### 2)

```python
mean_vector = df_num[['Revenue', 'Cost']].mean().values
a = np.array([1, -1]) 
mean_net_profit = a @ mean_vector
print("Mean vector [Revenue, Cost]:", mean_vector)
print("Mean net profit (linear transformation):", mean_net_profit)
print("Podemos ver que", mean_net_profit, "es muy similar a : ", mean_per_column['P&L'])
```

---
### 3)

```python
cov_matrix = df[['Revenue', 'Cost']].cov()
print("Sample covariance matrix:")
print(cov_matrix)
```

---
### 4)

```python
S = df[['Revenue', 'Cost']].cov().values
a = np.array([1, -1])
var_net_profit = a.T @ S @ a
print("Sample variance of net profit:", var_net_profit)
```

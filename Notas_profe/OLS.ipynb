# Multiple Regression and Multivariate Time Series

## Ordinary Least Squares. Assumptions of the Multiple Regression Model.

---

# 1. Motivation

Multiple linear regression is the foundational model of statistical learning,
econometrics, and much of modern machine learning.

The Ordinary Least Squares (OLS) estimator is not merely a computational method.
It is:

- A geometric projection,
- A solution to a quadratic optimization problem,
- A maximum likelihood estimator under Gaussian noise,
- The Best Linear Unbiased Estimator (Gauss–Markov),
- The prototype of regularized and high-dimensional methods.

---

# 2. The Population Model

For observations $i = 1, \dots, n$:

$$
y_i = \beta_0 + \beta_1 x_{i1} + \dots + \beta_p x_{ip} + \varepsilon_i
$$

Matrix notation:

$$
\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}
$$

Where:

- $\mathbf{y} \in \mathbb{R}^{n \times 1}$
- $\mathbf{X} \in \mathbb{R}^{n \times (p+1)}$
- $\boldsymbol{\beta} \in \mathbb{R}^{(p+1) \times 1}$
- $\boldsymbol{\varepsilon} \in \mathbb{R}^{n \times 1}$

We assume $n > p+1$.

---

# 3. The OLS Optimization Problem

We define the objective:

$$
\min_{\beta} S(\beta)
$$

where

$$
S(\beta) =
(\mathbf{y} - \mathbf{X}\beta)^T
(\mathbf{y} - \mathbf{X}\beta)
$$

This is a strictly convex quadratic form if $X^TX$ is positive definite.

---

# 4. Derivation via Matrix Calculus

Expand:

$$
S(\beta)
= \mathbf{y}^T\mathbf{y}
- 2\beta^T \mathbf{X}^T \mathbf{y}
+ \beta^T \mathbf{X}^T \mathbf{X} \beta
$$

Gradient:

$$
\nabla_\beta S(\beta)
=
-2\mathbf{X}^T\mathbf{y}
+ 2\mathbf{X}^T\mathbf{X}\beta
$$

Set equal to zero:

$$
\mathbf{X}^T\mathbf{X}\hat{\beta}
=
\mathbf{X}^T\mathbf{y}
$$

These are the **normal equations**.

If $X^TX$ is invertible:

$$
\hat{\beta}
=
(\mathbf{X}^T\mathbf{X})^{-1}
\mathbf{X}^T\mathbf{y}
$$

---

# 5. Geometric Interpretation

Let $\mathcal{C}(X)$ denote the column space of $X$.

OLS finds $\hat{y} \in \mathcal{C}(X)$ such that:

$$
\|\mathbf{y} - \hat{\mathbf{y}}\|^2
$$

is minimized.

Thus:

$$
\hat{\mathbf{y}} = P_X \mathbf{y}
$$

where

$$
P_X = X(X^TX)^{-1}X^T
$$

Properties:

1. Symmetric: $P_X^T = P_X$
2. Idempotent: $P_X^2 = P_X$

Residual vector:

$$
\hat{\varepsilon} = y - \hat{y}
$$

Orthogonality:

$$
X^T \hat{\varepsilon} = 0
$$

Thus residuals are orthogonal to the column space of $X$.

---

# 6. Assumptions of the Classical Linear Model

## A1: Linearity in Parameters

The model is linear in $\beta$.

---

## A2: Full Column Rank

$$
rank(X) = p+1
$$

---

## A3: Strict Exogeneity

$$
E[\varepsilon \mid X] = 0
$$

This ensures:

$$
E[\hat{\beta}] = \beta
$$

---

## A4: Homoskedasticity

$$
Var(\varepsilon \mid X) = \sigma^2 I
$$

---

## A5: No Autocorrelation

$$
Cov(\varepsilon_i, \varepsilon_j) = 0
$$

---

## A6: Normality (optional for exact inference)

$$
\varepsilon \sim \mathcal{N}(0, \sigma^2 I)
$$

---

# 7. Unbiasedness of OLS

Substitute model into estimator:

$$
\hat{\beta}
=
(X^TX)^{-1}X^T(X\beta + \varepsilon)
$$

$$
=
\beta
+
(X^TX)^{-1}X^T\varepsilon
$$

Take expectation conditional on $X$:

$$
E[\hat{\beta} \mid X]
=
\beta
+
(X^TX)^{-1}X^T E[\varepsilon \mid X]
$$

Under exogeneity:

$$
E[\hat{\beta} \mid X] = \beta
$$

Thus OLS is unbiased.

---

# 8. Variance of the Estimator

From:

$$
\hat{\beta}
=
\beta
+
(X^TX)^{-1}X^T\varepsilon
$$

We compute:

$$
Var(\hat{\beta} \mid X)
=
(X^TX)^{-1}X^T
Var(\varepsilon \mid X)
X(X^TX)^{-1}
$$

Under homoskedasticity:

$$
Var(\hat{\beta})
=
\sigma^2 (X^TX)^{-1}
$$

---

# 9. Gauss–Markov Theorem (Proof Sketch)

Let $\tilde{\beta} = A y$ be any linear unbiased estimator.

Unbiasedness implies:

$$
AX = I
$$

Variance:

$$
Var(\tilde{\beta}) = \sigma^2 A A^T
$$

One can show:

$$
Var(\tilde{\beta}) - Var(\hat{\beta})
$$

is positive semi-definite.

Thus OLS has minimum variance among all linear unbiased estimators.

---

# 10. Sampling Distribution

If normality holds:

$$
\hat{\beta} \sim
\mathcal{N}
\left(
\beta,
\sigma^2 (X^TX)^{-1}
\right)
$$

Define:

$$
\hat{\sigma}^2 =
\frac{\hat{\varepsilon}^T\hat{\varepsilon}}
{n - p - 1}
$$

Then:

$$
\frac{\hat{\beta}_j - \beta_j}
{\sqrt{\hat{\sigma}^2 c_{jj}}}
\sim t_{n-p-1}
$$

---

# 11. Sum of Squares Decomposition

Define:

$$
SST = \|y - \bar{y}\|^2
$$

$$
SSE = \|y - \hat{y}\|^2
$$

$$
SSR = \|\hat{y} - \bar{y}\|^2
$$

Then:

$$
SST = SSR + SSE
$$

This follows from orthogonality of residuals and fitted values.

---

# 12. Maximum Likelihood Interpretation

If:

$$
\varepsilon \sim \mathcal{N}(0, \sigma^2 I)
$$

The likelihood:

$$
L(\beta, \sigma^2)
\propto
\exp
\left(
-\frac{1}{2\sigma^2}
(y - X\beta)^T
(y - X\beta)
\right)
$$

Maximizing likelihood in $\beta$ is equivalent to minimizing
sum of squared residuals.

Thus OLS = MLE under Gaussian noise.

---

# 13. Asymptotic Properties

Under mild regularity conditions:

$$
\sqrt{n}(\hat{\beta} - \beta)
\to_d
\mathcal{N}(0, Q^{-1}\Sigma Q^{-1})
$$

where:

$$
Q = \lim \frac{1}{n} X^TX
$$

Even without normality, OLS is asymptotically normal.

---

# 14. Heteroskedasticity

If:

$$
Var(\varepsilon \mid X) = \Omega
$$

Then OLS remains unbiased but inefficient.

Robust covariance estimator:

$$
\hat{V}_{rob}
=
(X^TX)^{-1}
\left(
\sum \hat{\varepsilon}_i^2 x_i x_i^T
\right)
(X^TX)^{-1}
$$

---

# 15. Multicollinearity

If $X^TX$ is nearly singular:

- Variance of $\hat{\beta}$ explodes.
- Estimates become unstable.
- Condition number becomes large.

Use SVD:

$$
X = U D V^T
$$

Then:

$$
\hat{\beta} = V D^{-1} U^T y
$$

Numerically more stable than normal equations.

---

# 16. When $X^TX$ is Singular

If $rank(X) < p+1$:

- Infinite solutions exist.
- Moore–Penrose pseudoinverse provides:

$$
\hat{\beta} = X^+ y
$$

This is the minimum-norm solution.

---

# 17. Omitted Variable Bias

Suppose true model:

$$
y = X\beta + Z\gamma + \varepsilon
$$

If $Z$ omitted:

$$
E[\hat{\beta}]
=
\beta
+
(X^TX)^{-1}X^TZ\gamma
$$

Bias occurs if $Cov(X,Z) \neq 0$.

---

# 18. Endogeneity

If:

$$
Cov(X,\varepsilon) \neq 0
$$

Then:

$$
E[\hat{\beta}] \neq \beta
$$

Solutions:

- Instrumental Variables
- Two-Stage Least Squares (2SLS)

---

# 19. Relation to Modern Machine Learning

Ridge regression:

$$
\hat{\beta}_{ridge}
=
(X^TX + \lambda I)^{-1}X^Ty
$$

Lasso:

$$
\min_\beta
\|y - X\beta\|^2 + \lambda \|\beta\|_1
$$

Thus OLS is the $\lambda=0$ case of regularization.

# Time Series Models Estimated with OLS
# 1. Trend Model

A trend model explains the evolution of a variable over time using a deterministic time index.

## Model

$$
y_t = \beta_0 + \beta_1 t + \varepsilon_t
$$

where:

- $t$ = time index
- $\beta_0$ = intercept
- $\beta_1$ = deterministic trend
- $\varepsilon_t$ = error term

## Matrix Form

$$
Y =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_T
\end{bmatrix},
\quad
X =
\begin{bmatrix}
1 & 1 \\
1 & 2 \\
\vdots & \vdots \\
1 & T
\end{bmatrix}
$$

OLS estimator:

$$
\hat{\beta} = (X'X)^{-1}X'Y
$$


---

# 2. Autoregressive Model AR(p)

An AR(p) model explains $y_t$ using its own past values.

## Model

$$
y_t = \beta_0 + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \varepsilon_t
$$

This model captures persistence over time.

## Matrix Form (Example AR(2))

$$
Y =
\begin{bmatrix}
y_3 \\
y_4 \\
\vdots \\
y_T
\end{bmatrix}
$$

$$
X =
\begin{bmatrix}
1 & y_2 & y_1 \\
1 & y_3 & y_2 \\
\vdots & \vdots & \vdots \\
1 & y_{T-1} & y_{T-2}
\end{bmatrix}
$$

OLS estimator:

$$
\hat{\beta} = (X'X)^{-1}X'Y
$$

OLS works because the model is linear in parameters and regressors are observed.

---

# 3. Vector Autoregression VAR(p)

A VAR model extends AR models to multiple time series.

Suppose we have two variables:

$$
\mathbf{y}_t =
\begin{bmatrix}
y_{1t} \\
y_{2t}
\end{bmatrix}
$$

## Model (VAR(1))

$$
\mathbf{y}_t = A_1 \mathbf{y}_{t-1} + \varepsilon_t
$$

where

$$
A_1 =
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
$$

This gives two equations:

$$
y_{1t} = a_{11}y_{1,t-1} + a_{12}y_{2,t-1} + \varepsilon_{1t}
$$

$$
y_{2t} = a_{21}y_{1,t-1} + a_{22}y_{2,t-1} + \varepsilon_{2t}
$$

## Estimation

Each equation is linear and can be estimated separately using OLS:

$$
\hat{\beta}_i = (X'X)^{-1}X'Y_i
$$

OLS is consistent if:

- The process is stationary
- Regressors are exogenous
- Errors are not serially correlated beyond the model structur
